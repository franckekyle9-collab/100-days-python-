```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
x = np.linspace(0, 10, n+1)
y = np.exp(-x/10) * np.sin(x)
plt.plot(x, y) # type: ignore 
```

```python
np.mean(y[(x>=4)*(x<=7)])
np.std(y[(x>=4)*(x<=7)])
```

```python
np.percentile(y[(x>=4)*(x<=7)], 80)
```

```python
plt.plot(x, np.gradient(y, x)) # type: ignore
```
