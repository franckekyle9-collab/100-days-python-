```python
import numpy as np  # type: ignore
# coefficients of the equation x**2 + 4x - 5 = 0
a ,b ,c = 1, 4, -5  # type: ignore
# calculate the discriminant
d = b**2 - 4*a*c  # type: ignore
# find two solutions
sol1 = (-b - np.sqrt(d)) / (2*a)  # type: ignore
sol2 = (-b + np.sqrt(d)) / (2*a)  # type:ignore
print(f"The solutions are {sol1} and {sol2}")  # type: ignore

```
